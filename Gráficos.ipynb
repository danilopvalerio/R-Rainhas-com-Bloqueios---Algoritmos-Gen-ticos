```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dados da tabela
data = {
    "Valor N": [8, 16, 32, 128, 256, 512],
    "Tempo estimado (ms)": [51, 1282, 4887, 93468, 616284, 1495808],
    "Gerao soluo": [1, 35, 72, 354, 1194, 1199]

}

df = pd.DataFrame(data)

# Configurao do estilo
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Grfico 1: Tempo de Execuo vs Valor de N (escala logartmica)
plt.subplot(1, 2, 1)
sns.lineplot(x="Valor N", y="Tempo estimado (ms)", data=df, marker="o", color="b")
plt.yscale("log")  # Escala logartmica para melhor visualizao
plt.title("Tempo de Execuo por Valor de N (log)")
plt.xlabel("Valor de N")
plt.ylabel("Tempo (ms) - Escala Log")

# Grfico 2: Geraes vs Valor de N
plt.subplot(1, 2, 2)
sns.barplot(x="Valor N", y="Gerao soluo", data=df, color="r")
plt.title("Nmero de Geraes por Valor de N")
plt.xlabel("Valor de N")
plt.ylabel("Geraes")

plt.tight_layout()
plt.show()
```
