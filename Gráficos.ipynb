```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dados da tabela
data = {
    "Valor N": [8, 16, 32, 128, 256, 512],
    "Tempo estimado (ms)": [51, 1282, 4887, 93468, 616284, 1495808],
    "Geração solução": [1, 35, 72, 354, 1194, 1199]

}

df = pd.DataFrame(data)

# Configuração do estilo
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Gráfico 1: Tempo de Execução vs Valor de N (escala logarítmica)
plt.subplot(1, 2, 1)
sns.lineplot(x="Valor N", y="Tempo estimado (ms)", data=df, marker="o", color="b")
plt.yscale("log")  # Escala logarítmica para melhor visualização
plt.title("Tempo de Execução por Valor de N (log)")
plt.xlabel("Valor de N")
plt.ylabel("Tempo (ms) - Escala Log")

# Gráfico 2: Gerações vs Valor de N
plt.subplot(1, 2, 2)
sns.barplot(x="Valor N", y="Geração solução", data=df, color="r")
plt.title("Número de Gerações por Valor de N")
plt.xlabel("Valor de N")
plt.ylabel("Gerações")

plt.tight_layout()
plt.show()
```
